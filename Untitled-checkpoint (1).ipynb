```python
import pandas as pd

# Creating a Python dictionary with the required columns
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Region': ['East', 'West', 'North', 'South', 'East'],
    'Sales': [1200, 25, 45, 300, 85],
    'Quantity': [1, 2, 1, 2, 3]
}

# Creating the DataFrame
df_custom = pd.DataFrame(data)

# Displaying the DataFrame
df_custom
import pandas as pd

# Loading the dataset
df = pd.read_csv("Retail_Dataset.csv")
# 1. Display the first 5 rows of the dataset
print("--- First 5 Rows ---")
display(df.head())

# 2. Display the last 5 rows
print("\n--- Last 5 Rows ---")
display(df.tail())

# 3. Show the structure of the dataset using info()
print("\n--- Dataset Structure ---")
df.info()

# 4. Display the shape of the dataset
print("\n--- Dataset Shape ---")
print(df.shape)
# Selecting and displaying only Product, Region, and Sales columns
selected_columns = df[['Product', 'Region', 'Sales']]
selected_columns
# 1. Display the row with index 10
print("--- Row with index 10 ---")
display(df.loc[10])

# 2. Display the Sales value of the row with index 20
print("\n--- Sales value of the row with index 20 ---")
sales_val_loc = df.loc[20, 'Sales']
print(sales_val_loc)
# 1. The first 6 rows and first 4 columns
print("--- First 6 rows and first 4 columns ---")
display(df.iloc[:6, :4])

# 2. The Sales value from the 3rd row (index 2 since Python is 0-indexed)
# Assuming 'Sales' column position. If Sales is the 3rd column, index is 2. 
# Alternatively, we find its column index dynamically:
sales_col_index = df.columns.get_loc('Sales')
sales_val_iloc = df.iloc[2, sales_col_index]

print("\n--- Sales value from the 3rd row ---")
print(sales_val_iloc)
# 1. The first 10 rows of the dataset
print("--- First 10 rows ---")
display(df.iloc[:10])  # or df.head(10)

# 2. Rows 15 to 25
print("\n--- Rows 15 to 25 ---")
display(df.iloc[15:26])
# 1. Rows where Region = "East"
print("--- Rows where Region is East ---")
east_region = df[df['Region'] == 'East']
display(east_region)

# 2. Rows where Sales > 1500
print("\n--- Rows where Sales > 1500 ---")
high_sales = df[df['Sales'] > 1500]
display(high_sales)
# Filtering rows where Region is West AND Sales is greater than 1200
filtered_multi = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]
filtered_multi
# Generate summary statistics
summary_stats = df.describe()
display(summary_stats)

# Identifying specific metrics from the description
# (Assumes standard column names match the description index output)
print("\n--- Identified Metrics ---")
print(f"Mean sales value: {summary_stats.loc['mean', 'Sales']}")
print(f"Maximum quantity sold: {summary_stats.loc['max', 'Quantity']}")
print(f"Minimum sales value: {summary_stats.loc['min', 'Sales']}")
# Calculating aggregation metrics using the Sales column
total_sales = df['Sales'].sum()
average_sales = df['Sales'].mean()
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()

# Displaying results
print(f"Total sales:   {total_sales}")
print(f"Average sales: {average_sales}")
print(f"Maximum sales: {max_sales}")
print(f"Minimum sales: {min_sales}")
r"C:\Users\yoges\Downloads\Retail_dataset.csv"
```
